```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import requests
import time
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
```

```python
driver = webdriver.Chrome(r'C:\Users\sawan\Desktop\chromedriver.exe')
```

- Have scrapped all the product url given in the below list

- laptops, Phones, Headphones, smart watches, Professional Cameras, Printers, Monitors and Router

```python
urll = 'https://www.flipkart.com/search?q=watches&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&sort=popularity'
driver.get(urll)

urls1 = [] #Get Pages URL (total 10 pages for each product)
urls=driver.find_elements(By.XPATH,'//a[@class="ge-49M"]')
for i in urls:
    urls1.append(i.get_attribute("href"))     


url_urls=[] # Collected Product URLs
for urlss in urls1:
    driver.get(urlss)
    urls2=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for i in urls2:
        url_urls.append(i.get_attribute("href"))     

    
More_reviews = [] # Collected reveiw pages (total 10 pages for each product)
for url in url_urls:
    driver.get(url)
    time.sleep(1)
    
    try:
        more_reviews = driver.find_element(By.XPATH,'//div[@class="_3UAT2v _33R3aa"]')
        more_reviews.click()
        time.sleep(1)
    except NoSuchElementException:
        continue
     

    try:
        Mreviews =driver.find_elements(By.XPATH,'//a[@class="ge-49M"]')
        for i in Mreviews:
            More_reviews.append(i.get_attribute("href"))    
    except NoSuchElementException:
        continue  
```

```python
print(len(More_reviews))
print(len(url_urls))
```

```python
heading8 = [] # collected info from each review page
heading9 = [] # collected info from each review page
```

```python
counter = []
for url in More_reviews[1718:2218]:
    driver.get(url)
    time.sleep(2)
    try:
        column9 = driver.find_elements(By.XPATH,'//div[@class="col _2wzgFH K0kLPL _1QgsS5"]')    
        for z in column9:
            heading9.append(z.text) 
    except NoSuchElementException:
        heading9.append('-') 
            
    try:
        column8 = driver.find_elements(By.XPATH,"//div[@class='_6K-7Co']")    
        for h in column8:
            heading8.append(h.text)
    except NoSuchElementException:
        heading8.append('-')
        
    counter.append(url)
    print('URL Scrapping Done -', counter.index(url))
```
